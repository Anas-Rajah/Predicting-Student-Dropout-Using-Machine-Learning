# Data Modelling 

```python
student_data = student_data.drop(student_data[student_data['Target']=='Enrolled'].index)
student_data.head()
```

```python
encoder = LabelEncoder()
```

```python
student_data['Target'] = encoder.fit_transform(student_data['Target'])
```

```python
student_data.head()
```

```python
# Set target variable as X and the remaining variables except Nationality and International as Y
X = student_data.drop(columns=['Nacionality','International','Target'], axis=1)
Y = student_data['Target']
```

```python
print(X, X.shape)
```

## Verileri EÄŸitim ve Test Verilerine BÃ¶lme
Makine Ã¶ÄŸrenimi modelimiz olarak lojistik regresyonla baÅŸlamak iÃ§in verileri eÄŸitim ve test verilerine bÃ¶ldÃ¼k. Verilerin %80'i eÄŸitim modelimiz, geri kalan %20'si ise test modeli olacaktÄ±r. Rastgele Ã¶rneklemenin Ã¼Ã§Ã¼ncÃ¼ durumunu seÃ§iyoruz.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## XGB Lojistik Regresyon
Lojistik regresyon artÄ±k regresyon modelleri iÃ§in kullanÄ±lan aÃ§Ä±k kaynak kitaplÄ±klardan biri olan Extreme Gradient Boosting (XGBoost) kullanÄ±larak uygulanacaktÄ±r. Bu durumda modelimiz iÃ§in 1000 n_estimators ile ikili lojistik ayarlanmÄ±ÅŸtÄ±r. N_tahmin ediciler, Ã¶zellik deÄŸiÅŸkenlerinden elde edilen verileri dikkate alarak karar aÄŸaÃ§larÄ±nÄ±n sayÄ±sÄ± veya sÄ±nÄ±flandÄ±rma gÃ¶revi gÃ¶rÃ¼r.

```python
bin_log = xgb.XGBClassifier(objective='binary:logistic', n_estimators=1000)
bin_log.fit(X_train, Y_train)
```

## Data Prediction and Evaluation of the Model
We now set the logistic regression model to the training data.

```python
target_prediction = bin_log.predict(X_test)
```

```python
print(target_prediction)
```

```python
data_accuracy = accuracy_score(Y_test, target_prediction)
```

```python
print("Accuracy:", data_accuracy)
```
